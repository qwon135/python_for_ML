## Polynomial Features
- 1차 다항식을 고차다항식으로 변경하는 기법  
    : x + y 를 x + y + xy + x^2 + y^2로 변경  
- sklearn.preprocessing.PolynomialFeatures 사용

### Dataset
- f 함수와 randn함수 f가 있다.
- 한정적은 데이터가 주어질때 linearegression을 사용해서 그래프를 그려낼 수 있을까?

```python
# 두가지 함수
def f(size):
    x = np.linspace(0, 5, size) # 0~5까지의 size개수만큼 균등하게 생성 (range개념)
    y = x * np.sin(x ** 2) + 1
    return (x,y)

def sample(size):
    x = np.linspace(0, 5, size)
    y = x * np.sin(x ** 2) + 1 + np.random.randn(x.size)*0.5
    return (x,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
X, y = sample(1000)
plt.scatter(X, y, s=3, c="black")
plt.show()
```

```python
# sklearn에는 2 dimension만 들어가므로 reshape한다.
X = X.reshape(-1,1)
y = y.reshape(-1,1)
X.shape, y.shape
```

```python
X[:10]
```

```python
from sklearn.linear_model import LinearRegression

# LinearRegression 생성후 fitting
lr = LinearRegression()
lr.fit(X,y)
```

degree를 계속 바꿔주는 이 과정을 parameter search라고 한다. Model이 크면 모든 항을 다 해볼 수는 없다.

### How to optimize
- RMSE의 최소값을 찾자
- Ridge, Lasso, LR모두 써보자
- Degree를 10~50까지 사용
- 결과를 한눈에 정리

### rmse

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

### greed search
- poly를 10~50가지 해본다.
- 실제로는 모든 수를 다 해볼 수 없어서 AWS로 10\~20 / 20\~30 / 30~40 등으롤 분할해서 Multi processing을 한다.

```python
poly_range = list(range(10, 50)) # 얼마나 poly_feature를 만들것인가?
rmse_lr_list = [] # Linear Regression
rmse_lasso_list = [] # Lasso
rmse_ridge_list = [] # Ridge

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

for poly_value in poly_range:
    poly_features = PolynomialFeatures(degree=poly_value) # degree는 poly_value에서 하나씩 사용
    X_poly = poly_features.fit_transform(X) # fitting
    lr = LinearRegression(fit_intercept = False) # intercept 를 False
    lr.fit(X_poly,y)# fitting
    
    # predit한 값을 각 rmse list에 저장
    
    # lr
    rmse_lr_list.append(rmse(lr.predict(X_poly), y)) 

    # Lasso
    lasso = Lasso()
    lasso.fit(X_poly,y)
    rmse_lasso_list.append(rmse(lasso.predict(X_poly), y))
    
    # Ridge
    ridge = Ridge()
    ridge.fit(X_poly,y)
    rmse_ridge_list.append(rmse(ridge.predict(X_poly), y))
```

### 예측치 보기(DataFrame)

```python
import pandas as pd
from pandas import DataFrame
data = {"poly_range":poly_range, "lr_rmse":rmse_lr_list, 
        "lasso_rmse":rmse_lasso_list,"ridge_rmse":rmse_ridge_list}
df = DataFrame(data).set_index("poly_range")
df
```

### Polymonial은 언제쓰나?
- 사용할 일이 많지는 않다.
- 한개의 변수가 Y값과 비선형적인 관계가 있다고 의심될 때
- 주기적인 패턴을 보이는 Series 데이터
- 모델 자체가 복잡해지면 해결가능한 부분이 많음 : SVM, Tree-based models 등

```python
plt.plot(poly_range, df["ridge_rmse"], label="ridge")
plt.plot(poly_range, df["lr_rmse"], label="lr")
plt.plot(poly_range, df["lasso_rmse"], label="lasso")
plt.legend()
plt.show()
```
